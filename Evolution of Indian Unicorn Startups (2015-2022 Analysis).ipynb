```python
# Importing important libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
%matplotlib inline
```

```python
data=pd.read_csv('Indian Unicorn startups 2023 updated.csv')
data
```

```python
data.describe()
```

# Number of unicorn startups each year

```python
Sector_counts = data['Sector'].value_counts()
Sector_counts
```

# Average Entry Valuation of unicorn startup


```python
avg_entry_val = data['Entry Valuation^^ ($B)'].mean()
print('Average Entry Valuation: $', avg_entry_val)
```

# Average Valuation of unicorn startup

```python
Avg_Val = data['Valuation ($B)'].mean()
print("Average Valuation: $", Avg_Val)
```

```python
# Check the data types of columns
data.dtypes
```

# Number of unique location

```python
unique_Location = data['Location'].nunique()
print('Number of Unique Location',unique_Location)
```

# Top 10 unicorn sectors


```python
top_10_Sectors = data['Sector'].value_counts().sort_values(ascending=False).head(10)
top_10_Sectors 
```

# Graphical Representation of top sectors with the most unicorn startups


```python
plt.figure(figsize=(25,15))
plt.xlabel('Sector', fontsize = 16)
plt.ylabel('Count', fontsize = 16)
plt.title('top sectors with the most unicorn startups', fontsize = 22)
plt.xticks(rotation=45,fontsize =12)
sns.barplot(top_10_Sectors.index,top_10_Sectors.values)
```

# Top 10 unicorn startups with the highest valuations


```python
top_valued_startups = data.nlargest(10, 'Valuation ($B)')
print(top_valued_startups[['Company', 'Valuation ($B)']])
```

# Graphical Representation of top 10 unicorn startups with the highest Valuations


```python
plt.figure(figsize=(25,15))
plt.xlabel('Company', fontsize = 16)
plt.ylabel('Valuation ($B)', fontsize = 16)
plt.title('top 10 unicorn startups with the highest Valuations', fontsize = 22)
plt.xticks(rotation=45,fontsize =12)
sns.barplot(x='Company',y='Valuation ($B)',data=top_valued_startups)
plt.show()
```

# Top 5 sectors with the highest total valuations


```python
top_5_valued_sectors = data.groupby('Sector')['Valuation ($B)'].sum().nlargest(5)
top_5_valued_sectors
```

# Distribution of most valued unicorn startups across sectors by Percentage


```python
plt.figure(figsize=(8,6))
top_5_valued_sectors.plot(kind = 'pie', autopct='%1.1f%%')
plt.title('Distribution of Most Valued Unicorn Startups Across Sectors')
plt.ylabel('')
plt.show()
```

# Total number of startups in each location


```python
Cities = data['Location'].value_counts().head(10)
Cities
```

# Graphical Representation of top 5 locations with the most unicorn startups


```python
plt.figure(figsize=(20,10))
plt.title('Number of Unicorn in each City',fontsize=29)
plt.ylabel('Cities',fontsize=25)
plt.xlabel('Count',fontsize=25)
sns.barplot(Cities.values,Cities.index,palette='hot')
plt.show()
```

```python
data.head()
```

# Total number of startups in each entry year


```python
unicorns_per_year = data['Entry'].value_counts().sort_index()
unicorns_per_year
```

# Bar chart for total number of startups in each entry year


```python
plt.plot(figsize=(30,10))
unicorns_per_year.plot(kind='bar', color='Blue')
plt.title('Total Number of Startups in Each Entry Year')
plt.xlabel('Entry Year')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45)
plt.show()
```

# Lets have a look at top industry from a closer point of view

# 1. E-Commerce

```python
Ecommerce = data[data['Sector']== 'E-Commerce']
Ecommerce
```

```python
Ecommerce.groupby('Company')['Valuation ($B)'].sum().sort_values(ascending=False).astype(int).reset_index()
```

# 2. Edtech

```python
ED_Tech = data[data['Sector']== 'Edtech']
ED_Tech
```

```python
ED_Tech.groupby('Company')['Valuation ($B)'].sum().sort_values(ascending=False).astype(int).reset_index()
```

# 3. Logistics

```python
Logistics = data[data['Sector']=='Logistics Services']
Logistics
```

```python
Logistics.groupby('Company')['Valuation ($B)'].sum().sort_values(ascending=False).astype(int).reset_index()
```

# Total Valuation by entry year


```python
total_valuation_per_year = data.groupby('Entry')['Valuation ($B)'].sum()
total_valuation_per_year
```

# Line chart for total valuation by entry year


```python
plt.figure(figsize=(20, 6))
total_valuation_per_year.plot(kind='line', marker='o', color='orange')
plt.title('Total Valuation by Entry Year')
plt.xlabel('Entry Year')
plt.ylabel('Total Valuation ($B)')
plt.xticks(rotation=45)
plt.show()
```

# Top 10 investors with the highest number of investments


```python
top_investors = data['Select Investors'].str.split(',', expand = True).stack().value_counts().head(10)
```

# Bar chart for top 10 investors with the highest number of investments


```python
plt.figure(figsize=(20,6))
sns.barplot(top_investors.index,top_investors.values)
plt.title('Top 10 investors with the highest number of investments')
plt.xlabel('Investor')
plt.ylabel('Number of Investments')
plt.xticks(rotation=45)
plt.show()

```
